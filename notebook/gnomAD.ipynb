```python
def head(file_dir, n, is_strip:bool = True):
    """head a file at n rows. like head -n in bash, will return a list of row, if strip, each row will do row.strip(strip)

    Args:
        file_dir (_type_): _description_
        n (_type_): _description_
        strip (Union[str, None], optional): _description_. Defaults to None.

    Returns:
        _type_: _description_
    """
    with open(file_dir, "r") as f:
        return [f.readline().strip() if is_strip else f.readline() for i in range(n)]
            
```
