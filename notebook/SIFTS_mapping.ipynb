```python
from copy import deepcopy
def explode(df:pd.DataFrame, column:str, sep:str = ","):
    tmp = deepcopy(df)  # shallow copy 

    tmp[column] = tmp[column].apply(lambda x: x.split(sep))
    return tmp.explode(column)
    
```

```python
add_df = SIFTS_mapping_df[SIFTS_mapping_df["Gene Names"].isna()] 
add_df["Gene Names"] = add_df["Entry Name"].apply(lambda x: x.split("_")[0])

tmp_df = pd.concat([SIFTS_mapping_df, add_df])

SIFTS_mapping_df_full = tmp_df.dropna(subset=["Gene Names"])
SIFTS_mapping_df_full
```

```python
SIFTS_mapping_df_full = explode(SIFTS_mapping_df_full, column="Gene Names", str = " ")  # 切分属于多个基因的Uniprot Accession
```
